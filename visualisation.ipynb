```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("DelayedFlights.csv")
```

```python
plt.figure(figsize=(10,6))
plt.title("Number of cancelled flights, by Month")

s = data.groupby("Month")["Cancelled"].sum()
sns.barplot(x=s.index.to_list(), y=s.values)
plt.ylabel("Cancelled flights count")
```

```python
plt.figure(figsize=(10,6))
plt.title("Number of cancelled flights, by Day of Month")

s = data.groupby("DayofMonth")["Cancelled"].sum()
sns.barplot(x=s.index.to_list(), y=s.values)
plt.ylabel("Cancelled flights count")
```

```python
plt.figure(figsize=(10,6))
plt.title("Number of cancelled flights, by Day of Week")

s = data.groupby("DayOfWeek")["Cancelled"].sum()
sns.barplot(x=s.index.to_list(), y=s.values)
plt.ylabel("Cancelled flights count")
```

```python
sns.lmplot(x="DepDelay", y="Distance", hue="Cancelled", data=data)
```
